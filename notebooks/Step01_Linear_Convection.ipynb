Step 01 - Linear Convection

Here, the goal was to understand the concept of discretization of space and time and the use of finite difference (forward and backward) methods for approximating partial differential equations. Then also transferring the simple algberaic approximate to python code.

```python
# importing libraries

import numpy as np
import matplotlib.pyplot as plt
import time, sys
```

```python
# initializing the parameters

nx = 41 #number of grid points
dx = 2 / (nx - 1) # grid spacing (we use 2 because the domain is from 0 to 2)
nt = 40 # number of time steps
dt = 0.025 # time step size
c = 1 # wave speed
```

```python
# Setting up the initial conditions (The initial velocity is given as u = 2 in the interval [0.5, 1] and u = 1 everywhere else in (0, 2) (i.e., a hat function).)

u = np.ones(nx) # initialize u as an array of ones
u[int(0.5 / dx):int(1 / dx + 1)] = 2 # set u = 2 between 0.5 and 1
print(u) # print the initial condition
```

```python
plt.plot(np.linspace(0, 2, nx), u) # ploting the initial condition
```

What i learnt: The Hat function gets to have more perfectly straight lines as the nuber of grid points is increased - This is due to 

```python
# Implementing the discretization of the linear convection equation using the finite difference method

un = np.ones(nx) # initialize un as a temporary array of ones

for n in range(nt): # iterate through time steps
    un = u.copy() # copy the existing values of u into un
    for i in range(1, nx): # iterate through grid points
        u[i] = un[i] - c * dt / dx * (un[i] - un[i - 1]) # update the value of u at each grid point
```

```python
plt.plot(np.linspace(0, 2, nx), u) # plot the final condition
```

What i learnt: The wave has now shifted, because what is being printed is the final condition of the quantity u (i.e after 25 time step or 0.625 secs) and the shape is longer a hat function (This shape changes with respect to the space discretization)
