Step 02 - Nonlinear Convection

Here, the goal is to understand the effect of a nonlinear term in the convection equation, in this case the wave speed/velocity.

```python
# Just like the first lesson, we import important libraries

import numpy as np
import matplotlib.pyplot as plt


# initializing the parameters

nx = 41 # number of grid points
dx = 2 / (nx - 1) # grid spacing (we use 2 because the domain is from 0 to 2)
nt = 20 # number of time steps
dt = 0.025 # time step size 


# creating an array to store the initial u values (boundary condition) 

u = numpy.ones(nx) # initialze u with every value equal to 1
u[int(.5 / dx) : int(1/dx + 1)] = 2 # making u = 2 at the given interval

# initializing a placeholder array to hold the value fo u for every timestep iteration

un = numpy.ones(nx) #initialized with ones
```

```python
# Implementing the discretization of the linear convection equation using the finite difference method

for n in range(nt): # iterate through time steps
    un = u.copy() # copy the existing values of u into un
    for i in range(1, nx): # iterate through grid points
        u[i] = un[i] - un[i] * dt / dx * (un[i] - un[i - 1]) # update the value of u at each grid point


plt.plot(np.linspace(0, 2, nx), u) # plot the final condition
```

What i Learnt: When the transport speed depends on the solution itself, the solution can change its own shape.
